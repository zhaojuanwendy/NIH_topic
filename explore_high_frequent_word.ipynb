```python
from gensim.corpora import Dictionary
tokenizer='spacy'
dictionary = Dictionary.load('data/corpus_ML_DL_{}/dictionary.gensim'.format(tokenizer))
```

```python
dictionary.dfs
```

```python
sorted_dicts = sorted(dictionary.dfs.items(),key=lambda item: item[1],reverse=True)
print(sorted_dicts[:50])
```

```python
temp = dictionary[0]  # This is only to "load" the dictionary.
high_fre_words = []
for p in sorted_dicts[:250]:
    idx = p[0]
    high_fre_words.append(dictionary.id2token.get(idx))
```

```python
",".join(w for w in high_fre_words)
```
