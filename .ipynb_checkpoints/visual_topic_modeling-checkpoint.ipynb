## evaluate topic

```python
def get_topic_coherences(num_topics_list,corpus,model_path):
    avg_topic_coherences = []
    for num_topics in num_topics_list:
#         print(num_topics)
        lda_model = gensim.models.ldamodel.LdaModel.load(path.join(model_path, 'lda-{}.model'.format(num_topics)),mmap='r')
        top_topics = lda_model.top_topics(corpus)
        avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics   
#         print(avg_topic_coherence)
        avg_topic_coherences.append(avg_topic_coherence)
    return avg_topic_coherences

def plot_coherence(model_subfolder=''):
    corpus = pickle.load(open(path.join('models/corpus_{}/corpus.pkl'.format(model_subfolder)), 'rb'))
    dictionary = Dictionary.load('models/corpus_{}/dictionary.gensim'.format(model_subfolder))
    model_path = 'models/{}/'.format(model_subfolder)
    num_topics_list = range(5, 60, 5)#,15,20,25
    avg_topic_coherences = get_topic_coherences(num_topics_list,corpus,model_path)
    ax = sns.lineplot(x=num_topics_list, y=avg_topic_coherences)
    plt.xlabel("topics number")
    plt.ylabel("coherence")
    plt.xticks(num_topics_list)
```

```python
plot_coherence('ML_DL_spacy')
```

```python
def get_perplexity(num_topics_list,corpus,model_path):
    avg_log_perplexity = []
    for num_topics in num_topics_list:
#         print(num_topics)
        lda_model = gensim.models.ldamodel.LdaModel.load(path.join(model_path, 'lda-{}.model'.format(num_topics)),mmap='r')
        avg_log_perplexity.append(lda_model.log_perplexity(corpus))
    return avg_log_perplexity

def plot_perplexity(tokenizer='spacy'):
    corpus = pickle.load(open(path.join('models/corpus_ML_DL_{}/corpus.pkl'.format(tokenizer)), 'rb'))
    dictionary = Dictionary.load('models/corpus_ML_DL_{}/dictionary.gensim'.format(tokenizer))
    model_path = 'models/ML_DL_{}/'.format(tokenizer)
    num_topics_list = range(5, 60, 5)#,15,20,25
    avg_log_perplexity = get_perplexity(num_topics_list,corpus,model_path)
    ax = sns.lineplot(x=num_topics_list, y=avg_log_perplexity)
    plt.xlabel("topics number")
    plt.xlabel("log_perplexity")
    plt.xticks(num_topics_list)
```

```python
plot_perplexity('spacy')
```

## visualize topics

```python
tokenizer = 'spacy'
corpus = pickle.load(open(path.join('models/corpus_ML_DL_{}/corpus.pkl'.format(tokenizer)), 'rb'))
dictionary = Dictionary.load('models/corpus_ML_DL_{}/dictionary.gensim'.format(tokenizer))
model_path = 'models/ML_DL_{}/'.format(tokenizer)
```

```python
import gensim
num_topics = 10
lda_model = gensim.models.ldamodel.LdaModel.load(path.join(model_path, 'lda-{}.model'.format(num_topics)))
topics = lda_model.show_topics(formatted=False, num_topics=num_topics)
```

```python
from pprint import pprint
# pprint(topics)
```

```python
import matplotlib.colors as mcolors
from wordcloud import WordCloud
cols = [color for name, color in mcolors.TABLEAU_COLORS.items()]  # more colors: 'mcolors.XKCD_COLORS'

if len(cols) < num_topics:
    cols = cols + cols[:(num_topics-len(cols))]

cloud = WordCloud(background_color='white',
                  width=2500,
                  height=1800,
                  max_words=10,
                  colormap='tab10',
                  color_func=lambda *args, **kwargs: cols[i],
                  prefer_horizontal=1.0)


# fig, axes = plt.subplots(round(num_topics/2), 2, figsize=(10, 10), sharex=True, sharey=True)
fig, axes = plt.subplots(round(num_topics/2), 2, figsize=(10, 10), sharex=True, sharey=True)

for i, ax in enumerate(axes.flatten()):
    fig.add_subplot(ax)
    topic_words = dict(topics[i][1])
    cloud.generate_from_frequencies(topic_words, max_font_size=300)
    plt.gca().imshow(cloud)
    plt.gca().set_title('Topic ' + str(i+1), fontdict=dict(size=16))
    plt.gca().axis('off')

plt.subplots_adjust(wspace=0, hspace=0)
plt.axis('off')
plt.margins(x=0, y=0)
plt.tight_layout()
```

```python
def print_top_words(num_topics, topics, n_top_words):
    for topic_idx in range(num_topics):
        topic_words = list(dict(topics[topic_idx][1]).keys())
        message = "Topic #%d: " % (topic_idx+1)
        message += ", ".join(topic_words)
        print(message)
    print()
    
print_top_words(num_topics, topics, n_top_words=30)
```

```python
top_topic_idxs =[]
docs = []

for doc_idx, doc_topics in enumerate(docs_to_topics):
  
    print(sorted(dict(doc_topics).items(),key=lambda item: item[1],reverse=True))
    ranked_doc_topics = sorted(dict(doc_topics).items(), key=lambda item: item[1], reverse=True)
#     print("Document #{}".format(doc_idx), ranked_doc_topics)
    if(len(ranked_doc_topics))>0:
        top_topic_idxs.append(ranked_doc_topics[0][0]+1) #index plus one
    else:
        top_topic_idxs.append(0)
```

```python
def get_assigned_topics(docs_to_topics, only_top_topic=True, num_topics=2):
    top_topic_idxs = []
    docs = []
    for doc_idx, doc_topics in enumerate(docs_to_topics):
        # sort the topics according to descended probability 
        ranked_doc_topics = sorted(dict(doc_topics).items(),key=lambda item: item[1], reverse=True) 
        assign_num_topics = 1 # assign only top ranked topic at default
        
        if not only_top_topic:        
            # if top topics prob<0.5, pick the num_topics 
            if ranked_doc_topics[0][1]<0.5:  
#                 print("Document #{}".format(doc_idx), ranked_doc_topics)
#                 print("Document #{}".format(doc_idx), ranked_doc_topics[0][1])
                if len(ranked_doc_topics)>=num_topics:
                    assign_num_topics = num_topics       
                else:
                    assign_num_topics = len(ranked_doc_topics)                
            else:
                assign_num_topics=1
        
        for i in range(0,assign_num_topics):
            top_topic_idxs.append(ranked_doc_topics[i][0]+1)
            docs.append(doc_idx)
            
    return top_topic_idxs, docs
```

```python
# get list of assgined topc idx and document idx
top_topic_idxs, doc_idxs = get_assigned_topics(docs_to_topics, only_top_topic=False,num_topics=2)
len(top_topic_idxs)
```
